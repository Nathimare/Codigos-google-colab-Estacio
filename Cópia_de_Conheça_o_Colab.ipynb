1. Desenvolva um algoritmo que permita ao usuário escolher entre duas possibilidades para conversão de
temperaturas. O algoritmo deve mostrar o menu abaixo, ler a opção de conversão e realizar o respectivo
cálculo de conversão. Observe as informações abaixo.
Menu de Conversão de Temperaturas



*   Conversão Celsius/Fahrenheit: F = 9/5 * C+32.
*   Conversão Fahrenheit/Celsius: C = 5/9 * (F-32).


Digite a opção desejada:
Obs.: Nas fórmulas, C representa a temperatura em graus Celsius e F representa a temperatura em graus
Fahrenheit.

```
print("------ Menu de Conversão de Temperaturas ----------")
print("Escolha C => Celsius/Fahrenheit!!!")
print("Escolha F => Fahrenheit/Celsius!!!")

opcao = input("Digite a Opção C (Celsius) ou F (Fahrenheit)!!!")

if (opcao == "C"):
  temperatura_c = float(input("Digite o grau em Celsius "))
  conversao = (9/5)*temperatura_c+32
  print(f"A conversão de Celsius/Fahrenheit = {conversao}")
elif (opcao == "F"):
  temperatura_f = float(input("Digite o grau em Fahrenheit "))
  conversao = (5/9)*(temperatura_f-32)
  print(f"A conversao de Fahrenheit/Celsius = {conversao}")
else:
  print("Opcao invalida, execute novamente o programa e forneca C ou F!!!")
```

1.Escreva um algoritmo que mostre o menu de opções a seguir, receba a opção do usuário e os
dados necessários para executar cada operação.
Menu:
1. Somar dois números
2. Multiplicação de dois números
3. Raiz quadrada de um número
4. Quadrado de um número

```
import math
print("------Calculadora------")
print("Menu:")

print("1. Somar dois números")
print("2. Multiplicação de dois números")
print("3. Raiz quadrada de um número")
print("4. Quadrado de um número")

opcao = int(input("Escolha uma das opções acima!: "))

if (opcao == 1):
  s1 = float(input("Digite o primeiro número: "))
  s2 = float(input("Agora digite o segundo número: "))
  soma = s1 + s2
  print(f"A soma de {s1} e de {s2} é: {soma}")

elif (opcao == 2):
  m1 = float(input("Digite o primeiro número: "))
  m2 = float(input("Agora digite o segundo número: "))
  mult = m1 * m2
  print(f"A Multiplicação de {m1} e {m2} é: {mult}")

elif (opcao == 3):
 r1 = float(input("Digite o número: "))
 raiz = math.sqrt(r1)
 print(f"A raiz de {r1} é: {raiz}")

elif (opcao == 4):
  q1 = float(input("Digite o número: "))
  qua = q1 ** 2
  print(f"O quadrado de {q1} é {qua}")

else:
  print("Digite um número válido!")
```

2.Sabendo-se que 100 quilowatts de energia custam um sétimo do salário mínimo, escreva um algoritmo
que receba o valor do salário mínimo, a quantidade de quilowatts gastos por uma residência e apresente as
seguintes informações:



*   o valor em reais de cada quilowatts;
*   o valor em reais a ser pago;
*   o valor a ser pago por essa residência com desconto de 10%.

```
sl = float(input("Digite o salário mínimo: "))
ql = float(input("Digite o quantidade do quilowatts: "))

ql_100Kw = sl / 7
ql_1Kw = ql_100Kw / 100
print(f"Valor de cada kWh: R$ {ql_1Kw:.2f}")

valor_total = ql_1Kw * ql
print(f"valor total a ser pago: R$ {valor_total:.2f}")

desconto = valor_total - (valor_total * 0.1)
print(f"Valor com 10% de desconto: R$ {desconto:.2f}")
```

Tipo|

    |      descrição          | Rendimento Mensal |

```
print("Menu:")
print("Os tipos de investimentos")
print("1.Poupança")
print("2.Fundos de renda fixa")
opcao = int(input("Escolha qual queira investir: "))

if opcao == 1:
  num = float(input("Quanto você quer investir? "))
  pou = num * 0.03
  print("Investindo R${}, no próximo mês você recebera R${:.2f}. Somando tudo fica: {:.2f}".format(num, pou, num + pou))

elif opcao == 2:
  num = float(input("Quanto você quer investir? "))
  fr = num * (1 + 0.04) ** 1/12
  print("Investindo R${}, no próximo mês você recebera R${:.2f}. Somando tudo fica: {:.2f}".format(num, fr, num + fr))

else:
  print("Escolha uma opção válida!")
```
